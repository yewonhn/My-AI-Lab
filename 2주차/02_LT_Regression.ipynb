```python
import pandas as pd
import numpy as np
from google.colab import drive

# 드라이브 연결
drive.mount('/content/drive')
path = '/content/drive/MyDrive/Colab Notebooks/diabetes.csv'
df = pd.read_csv(path)

# 회귀(Regression)를 위해 수치형 데이터인 'BMI'를 예측 목표(y)로 잡습니다.
y = df['BMI']
X = df.drop(['BMI', 'Outcome'], axis=1) # 예측 대상과 분류 레이블은 제외
display(X.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression # 선형 회귀
from sklearn.metrics import mean_squared_error, r2_score

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# 모델 생성 및 학습
model_lr = LinearRegression()
model_lr.fit(X_train, y_train)

# 예측
pred = model_lr.predict(X_test)

print("\n--- Linear Regression Results ---")
print(f"MSE (오차): {mean_squared_error(y_test, pred):.4f}")
print(f"R2 Score (결정계수): {r2_score(y_test, pred):.4f}")
# R2 Score가 1에 가까울수록 모델이 수치를 잘 맞힌다는 뜻
```
