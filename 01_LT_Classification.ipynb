```python
import pandas as pd
import numpy as np

# 데이터 경로 설정 (본인 드라이브 경로 확인!)
path = '/content/drive/MyDrive/Colab Notebooks/diabetes.csv'

# DataFrame으로 읽기 및 확인
df = pd.read_csv(path)
display(df.head())
y = df['Outcome']
print("클래스 분포:\n", y.value_counts())
```

```python
X = df.drop('Outcome', axis=1)
from sklearn.model_selection import train_test_split

# 학습 데이터와 테스트 데이터 분리 (75:25)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
print("학습 데이터 크기:", X_train.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

# 로지스틱 회귀 모델 생성 및 학습
clf_lr = LogisticRegression(random_state=0, max_iter=1000) # 에러 방지 max
clf_lr.fit(X_train, y_train)

# 예측 및 성능 평가
pred_lr = clf_lr.predict(X_test)

print("\n--- Logistic Regression Classifier Results ---")
print(f"정확도(Accuracy): {accuracy_score(y_test, pred_lr):.4f}")
print("혼동 행렬(Confusion Matrix):")
print(confusion_matrix(y_test, pred_lr))
```

```python
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score

from sklearn.metrics import confusion_matrix

clf_lr = LogisticRegression(random_state=0)

clf_lr.fit(X_train, y_train)

pred_lr = clf_lr.predict(X_test)

print ("\n--- Logistic Regression Classifier ---")

print (accuracy_score(y_test, pred_lr))

print (confusion_matrix(y_test, pred_lr))
```
