```python
import pandas as pd
import numpy as np  # 늘 있을 것.
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/heart.csv")
df
```

```python
X = df.drop('target',axis=1)  # 타겟 제외 나머지 컬럼
X.head()
```

```python
X = df.drop('age',axis=1)   # 나이 제외 나머지 컬럼
X.head()
```

```python
y = df['age']   # 나이만 떼서 붙임
y.head()
```

## 3. 훈련 데이터와 테스트 데이터로 분리

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(X,
                                                  y,
                                                  test_size=0.2,    # 테스트는 20% 분량 준다는 것.
                                                  shuffle=True,
                                                  random_state=12)
print(X_train.shape, y_train.shape)  # 훈련 데이터 수
print(X_test.shape, y_test.shape)    # 테스트 데이터 수
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.metrics import mean_squared_error
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
ly_preds = model.predict(X_test)

print('평균제곱근오차', mean_squared_error(ly_preds, y_test))

def mse_np(actual, predicted):
    return np.mean((np.array(actual) - np.array(predicted)) ** 2)

print('평균제곱근오차', mse_np(ly_preds, y_test))

def mse(actual, predicted):
    sum_square_error = sum((a - p) ** 2for a, p in zip(actual, predicted))
    mean_square_error = sum_square_error / len(actual)
    return mean_square_error

print('평균제곱근오차', mse(ly_preds, y_test))
```

```python
# 1. 모델 생성
lr_reg = LinearRegression()
dt_reg = DecisionTreeRegressor(random_state=0)
rf_reg = RandomForestRegressor(random_state=0)

# 2. 모델 학습
lr_reg.fit(X_train, y_train)
dt_reg.fit(X_train, y_train)
rf_reg.fit(X_train, y_train)

# 3. 예측 및 평가
models = [lr_reg, dt_reg, rf_reg]
for model in models:
    pred = model.predict(X_test)
    name = model.__class__.__name__
    mae = mean_absolute_error(y_test, pred)
    r2 = r2_score(y_test, pred)

    print(f"--- {name} ---")
    print(f"Mean Absolute Error (MAE): {mae:.2f}")
    print(f"R2 Score: {r2:.2f}\n")
```
