```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor  # Regressor로 변경
from sklearn.metrics import mean_squared_error, r2_score

path = '/content/drive/MyDrive/Colab Notebooks/wine.csv'
df = pd.read_csv(path)

# Wine은 지우고 Alcohol을 정답(y)으로 설정
y = df['Alcohol']
X = df.drop(['Wine', 'Alcohol'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# 모델 생성 및 학습
model = RandomForestRegressor(random_state=0)
model.fit(X_train, y_train)

# 결과 출력
pred = model.predict(X_test)
print(f"6번 알코올 예측 MSE: {mean_squared_error(y_test, pred):.4f}")
print(f"6번 알코올 예측 R2 Score: {r2_score(y_test, pred):.4f}")
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X_reg = df.drop(['Wine', 'Alcohol'],axis=1)
y_reg = df['Alcohol']

X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)
```

```python
reg_tree = DecisionTreeRegressor(max_depth=5, random_state=42)  # 모델 생성
reg_tree.fit(X_train_r, y_train_r)    # 이게 학습
```

```python
y_pred_r = reg_tree.predict(X_test_r)

# 모델 평가 하는 코드
print(f"오차 제곱 평균: {mean_squared_error(y_test_r, y_pred_r): .4f}")
print(f"설명력: {r2_score(y_test_r, y_pred_r): .4f}")
```
